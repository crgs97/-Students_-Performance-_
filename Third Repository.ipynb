# 3. Who scores the most on average for math, reading and writing based on;

```python
datasp['Math Score'].mean()
```

```python
plt.figure(figsize=(15,5))
plt.subplots_adjust(left=0.125, bottom=0.1, right=1.0, top=1.0,wspace=0.8, hspace=0.2)

plt.subplot(131)

plt.title('Math Scores')

sns.barplot(hue="Gender", y="Math Score", x="Test Preparation Course", data=datasp, palette='husl')
plt.axhline(y = 66.001, color = 'black', linestyle = 'dashdot', label = "Avg Math Score")
plt.legend(bbox_to_anchor = (1.34, 1), loc = 'upper center')


plt.subplot(132)

plt.title('Reading Scores')
sns.barplot (hue="Gender", y="Reading Score", x="Test Preparation Course", data=datasp)
plt.axhline(y = 69.195, color = 'black', linestyle = 'dashdot', label = "Avg Reading Score")
plt.legend(bbox_to_anchor = (1.34, 1), loc = 'upper center')

plt.subplot(133)

plt.title('Writing Scores')
sns.barplot (hue="Gender", y="Writing Score", x="Test Preparation Course", data=datasp, palette='coolwarm')
plt.axhline(y = 68.119, color = 'black', linestyle = 'dashdot', label = "Avg Writing Score")
plt.legend(bbox_to_anchor = (1.34, 1), loc = 'upper center')
plt.show()
```

Inference: The above Graph gives a decent analytics over the Respective Scores and the Test Preparation Course.

# 4. What do you think about the scoring variation for math, reading and writing based on

# â— Gender

```python
sns.set_palette('husl') 
plt.figure(figsize=(15,6))
sns.histplot(x='Math Score', data=datasp, kde=True, hue='Gender')
plt.title('Math Scores of Gender')
plt.show()

sns.set_palette('bright') 
plt.figure(figsize=(15,6))
sns.histplot(x='Reading Score', data=datasp, kde=True, hue='Gender')
plt.title('Reading Scores of Gender')
plt.show()

sns.set_palette('coolwarm') 
plt.figure(figsize=(15,6))
sns.histplot(x='Writing Score', data=datasp, kde=True, hue='Gender')
plt.title('Writing Scores of Gender')
plt.show()
```

Inference: The above Graph gives a decent analytics over the Respective Scores wrt Gender.
